```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("Student Performance.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["math percentage"], kde=False)
plt.title("Histogram of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["math percentage"], kde=True)
plt.title("Math Score Distribution with KDE")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.kdeplot(df["math percentage"], fill=True)
plt.title("KDE Plot of Math Scores")
plt.xlabel("Math Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["math percentage"])
plt.title("Boxplot of Math Scores")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="sex", y="math percentage", data=df)
plt.title("Math Score by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="math percentage", hue="sex", kde=True)
plt.title("Math Score Distribution by Gender")
plt.xlabel("Math Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="test preparation course", y="math percentage", data=df)
plt.title("Math Score vs Test Preparation")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="sex", y="math percentage", data=df)
plt.title("Math Score by Gender")
plt.savefig("math_score_by_gender.png")
plt.show()
```

The distribution of math scores appears moderately symmetric with most values concentrated between 60 and 80. The histogram and KDE plot show slight left skewness, while the boxplot reveals a few low-score outliers. Gender-based comparison indicates that male students have a marginally higher median score, though the interquartile ranges overlap, suggesting similar performance variability. Overall, the dataset reflects consistent academic performance with minor group-level differences.
